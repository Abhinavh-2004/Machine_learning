```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
data = load_iris()
X = data.data
Y = data.target

lda = LDA(n_components=1)
X_lda = lda.fit_transform(X,Y)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c = Y)
plt.scatter(X_lda, np.zeros_like(X_lda), c = Y)
plt.show()
```

```python
plt.scatter(X_lda,np.zeros_like(X_lda), c = Y)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Original Data
axs[0].scatter(X[:, 0], X[:, 1], c=Y, cmap='coolwarm', s=100)
axs[0].set_title("Original Data (First 2 Features)")
axs[0].set_xlabel("Feature 1")
axs[0].set_ylabel("Feature 2")

# Plot 2: LDA Projected (1D)
axs[1].scatter(X_lda, np.zeros_like(X_lda), c=Y, cmap='coolwarm', s=100)
axs[1].set_title("After LDA (1D Projection)")
axs[1].set_xlabel("Linear Discriminant")
axs[1].set_yticks([])

plt.tight_layout()
plt.show()
```
